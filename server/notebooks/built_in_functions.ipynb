```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.util import random_noise
from skimage import img_as_ubyte
```

```python
def gaussian(path):
    img = cv2.imread(path,0)
    gauss_noise = np.zeros(img.shape, dtype = np.uint8)
    cv2.randn(gauss_noise,128,20)
    gauss_noise = (gauss_noise*0.5).astype(np.uint8)
    gn_img=cv2.add(img,gauss_noise)
    noised = plt.imshow(gn_img,cmap='gray')
    plt.title("Gaussian noise")
    
    return noised
    
```

```python
gaussian("C:/Users/ameer/Downloads/WhatsApp Image 2023-03-07 at 10.57.20 PM.jpeg")
```

```python
def salt_and_pepper(path):
    img = cv2.imread(path,0)
    noise_img = random_noise(img, mode='s&p',amount=0.3)
    noise_img = np.array(255*noise_img, dtype = 'uint8')
    noised = plt.imshow(noise_img,cmap='gray')
    plt.title("Salt and pepper noise")
    
    return noised
    
```

```python
salt_and_pepper("C:/Users/ameer/Downloads/WhatsApp Image 2023-03-07 at 10.57.20 PM.jpeg")
```

```python
def uniform_noise(path):
    img = cv2.imread(path,0)
    img = img/255
    # uniform noise
    x, y = img.shape
    n = np.zeros((x,y), dtype=np.float64)
    for i in range(x):
        for j in range(y):
            n[i][j] = np.random.uniform(0,0.2)


    # add noise to image
    noise_img = img + n
    noise_img = np.clip(noise_img, 0, 1)
    noised = plt.imshow(noise_img,cmap='gray')
    plt.title("Uniform noise")
    
    return noised
    
```

```python
uniform_noise("C:/Users/ameer/Downloads/WhatsApp Image 2023-03-07 at 10.57.20 PM.jpeg")
```

```python
def mean_filter(path): 
    img= cv2.imread(path,0)
    blur=cv2.fastNlMeansDenoising(img, None, 30, 7, 21) 
    filtered = plt.imshow(blur,cmap='gray')
    plt.title("Mean filter")
    
    return filtered
```
